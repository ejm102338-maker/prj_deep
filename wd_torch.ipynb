```python
# 인공신경망 만드는 방법 (보통 클래스를 만들어서 사용)
import torch.nn as nn

class Net(nn.Module):
    def __init__(self):
        super(Net,self).__init__()
        print("-"*30)
        self.fc1 = nn.Linear(in_features=3,out_features=4,bias=True)
        self.fc1_act = nn.ReLU()
        self.fc2 = nn.Linear(in_features=4,out_features=3,bias=True)
        self.fc2_act = nn.ReLU()
        self.fc3 = nn.Linear(in_features=3,out_features=1)

    def forward(self,x): #순정파 함수
        out = self.fc1(x)
        out = self.fc1_act(out)
        out = self.fc2(out)
        out = self.fc2_act(out)
        out = self.fc3(out)
        return out

model1 = Net()
model2 = Net()
model3 = Net()
```

```python
X = torch.randn(100,1) # 랜덤한값으로 행100개 열1개
X = X * 10
X.shape
```

```python
y = X+torch.randn(100,1)
y.shape
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X,y)
```

```python
import torch.nn as nn

class LRModel(nn.Module):
    def __init__(self):
        super(LRModel,self).__init__()
        self.fc1 = nn.Linear(in_features=1,out_features=4,bias=True)
        self.fc1_act = nn.ReLU()
        self.fc2 = nn.Linear(in_features=4,out_features=3,bias=True)
        self.fc2_act = nn.ReLU()
        self.fc3 = nn.Linear(in_features=3,out_features=1)

    def forward(self,x):
        x = self.fc1(x)
        x = self.fc1_act(x)
        x = self.fc2(x)
        x = self.fc2_act(x)
        x = self.fc3(x)
        return x

model = LRModel()
model
```

```python
list(model.parameters())
```

```python
import torch.optim as optim # 최적화 함수

criterion = nn.MSELoss() # 손실함수 (회귀:MSE : 평균오차제곱)
optimizer = optim.Adam(model.parameters(),lr=0.005) # 최적화 함수 (가중치(w)값을 옮긴다),lr:왼쪽으로 얼만큼 가야하는지 결정 0.001->1e-3 보통 0.001~0.003사용
epochs = 3000

loss_list = []
for epoch in range(epochs):
    optimizer.zero_grad()
    pred = model(X) # 예측값 계산
    loss = criterion(pred,y) # 손실값 계산

    loss_list.append(loss.item())
    print(f"Loss{epoch} : {loss}")

    # 손실함수 미분 : 역전파
    loss.backward()

    # 가중치 업데이트
    optimizer.step()
```

```python
plt.plot(range(epochs),loss_list)
plt.xlabel("epoch")
plt.ylabel("loss")
```
